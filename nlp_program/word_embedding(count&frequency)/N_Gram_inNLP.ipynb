````python
import nltk 
import re 
from nltk.tokenize import word_tokenize
from nltk.corpus import stopwords
from nltk.stem import WordNetLemmatizer
````

````python
lemma = WordNetLemmatizer()
lemma
````

````python
corpus = []

for i in range(0,len(message)):
    review = re.sub('[^a-zA-Z]', ' ', str(message.iloc[i, 1]))
    review = review.lower()
    review = review.split()
    review = [lemma.lemmatize(word) for word in review if not word in stopwords.words('english')]
    review = ' '.join(review)
    corpus.append(review)
    
````

# **N-Grams in NLP**

## **What are N-Grams?**
- **N-Grams** are contiguous sequences of `N` words from a given text.
- Used in **text processing, NLP, and machine learning** for tasks like **text prediction, spam detection, and speech recognition**.

## **Types of N-Grams**
| N | Name | Example (for "I love NLP") |
|---|------|----------------------------|
| 1 | **Unigram** | "I", "love", "NLP" |
| 2 | **Bigram** | "I love", "love NLP" |
| 3 | **Trigram** | "I love NLP" |
| N | **N-Gram** | "I love NLP models" (for N=4) |

## **Example: Generating N-Grams in Python**
```python
from nltk.util import ngrams
from nltk.tokenize import word_tokenize

text = "I love learning NLP"
tokens = word_tokenize(text)

bigrams = list(ngrams(tokens, 2))  # Generate bigrams
trigrams = list(ngrams(tokens, 3))  # Generate trigrams

print("Bigrams:", bigrams)
print("Trigrams:", trigrams)


````python
cv = CountVectorizer(max_features = 15 , binary = True,ngram_range=(1,3))

y = cv.fit_transform(corpus).toarray()

cv
````
